# Pre processing

```python
sd_train.drop(['Medicine_ID'], axis = 1, inplace = True) #on supprime la colonne d'identifiants ( 1557 valeurs uniques donc risque d'overfitting dans l'apprentissage)
```

```python
sd_test.drop(['Medicine_ID'], axis = 1, inplace = True)
```

```python
sd_train.head()
```

```python
sd_train_encoded = pd.get_dummies(sd_train,
columns=['DistArea_ID','Medicine_Type','SidEffect_Level','Area_Type','Area_City_Type','Area_dist_level']) #encodage one hot pour les variables cat√©gorielles
```

```python
sd_test_encoded = pd.get_dummies(sd_test,
columns=['DistArea_ID','Medicine_Type','SidEffect_Level','Area_Type','Area_City_Type','Area_dist_level'])
```

```python
sd_train_encoded.shape
```

```python
sd_test_encoded.shape
```

```python
# On essaie diff√©rentes strategy d'imputing ( pas de diff√©rence notable finalement m√™me si le KNN est donne un meilleur mae par la suite)
from sklearn.impute import SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer

imp0 = SimpleImputer(strategy='median')
imp1 = IterativeImputer(max_iter=30, random_state=0)
imp2 = KNNImputer(n_neighbors=10, weights="uniform",)
```

```python
sd_train_impute0 = pd.DataFrame(imp0.fit_transform(sd_train_encoded), columns = sd_train_encoded.columns) #on impute
sd_train_impute1 = pd.DataFrame(imp1.fit_transform(sd_train_encoded), columns = sd_train_encoded.columns)
sd_train_impute2 = pd.DataFrame(imp2.fit_transform(sd_train_encoded), columns = sd_train_encoded.columns)
```

```python
sd_test_impute2 = pd.DataFrame(imp2.fit_transform(sd_test_encoded), columns = sd_test_encoded.columns)
```

```python
sd_train_impute0.Counterfeit_Weight
```

```python
sd_train_impute1.Counterfeit_Weight
```

```python
sd_train_impute2.Counterfeit_Weight
```

# test & train set

```python
from sklearn.model_selection import train_test_split
```

```python
sd_train1,sd_train2=train_test_split(sd_train_impute2,test_size=0.2,random_state=2)
```

```python
x_train1=sd_train1.drop('Counterfeit_Sales',axis=1)

y_train1=sd_train1['Counterfeit_Sales']
```

```python
x_train1.shape
```

# linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train1,y_train1)  #entrainement sur un mod√®le simple
```

```python
lm.intercept_
```

```python
list(zip(x_train1.columns,lm.coef_))  #on fait appara√Ætre le coeff ( le poids) de chaque feature dans 
                                        #le mod√®le de regression
```

```python
x_train2=sd_train2.drop('Counterfeit_Sales',axis=1)
```

```python
predicted_ir=lm.predict(x_train2)
```

```python
print(predicted_ir)
```

```python
#on v√©rifie la performance sur le set de validation en choisissant la m√©trique mae

from sklearn.metrics import mean_absolute_error

ytest=sd_train2['Counterfeit_Sales']
mae=mean_absolute_error(ytest,predicted_ir)
print(mae)
```

```python
from sklearn.metrics import median_absolute_error
ytest=sd_train2['Counterfeit_Sales'] #on v√©rifie la performance sur le set de validation avec la m√©trique median absolute error
                                    #la medae est moins "punitive" pour les valeurs extr√™mes ( qui s'√©loignent de la moyenne)
medae=median_absolute_error(ytest,predicted_ir)
print(medae)
```

# Ridge  Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
lambdas=np.linspace(1,100,1000)
```

```python
params={'alpha':lambdas}    
```

```python
model=Ridge(fit_intercept=True) 
```

```python
grid_search=GridSearchCV(model,param_grid=params,cv=50,scoring='neg_mean_absolute_error') 

#On recherche les meilleurs hyperparam√®tres sur le crit√®re de la mae
```

```python
grid_search.fit(x_train1,y_train1)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_score_
```

```python
test_pred1 = grid_search.predict(sd_test_impute2)  # pr√©diction sur le test set
test_pred1
```
